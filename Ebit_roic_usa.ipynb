```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
driver = webdriver.Chrome()
url = 'https://stockanalysis.com/stocks/screener/'
driver.get(url)
```

```python
local_tabela = '/html/body/div/div[1]/div[2]/main/div[3]/div[2]/table'
elemento = driver.find_element("xpath", local_tabela)
html_tabela = elemento.get_attribute('outerHTML')
tabela = pd.read_html(str(html_tabela), thousands = '.', decimal = ',')[0]

tabela = tabela.set_index("Symbol")
```

```python
tabela.count()
```

```python
tabela = tabela[['Stock Price', 'EV/EBIT', 'ROIC', 'Volume', "Market Cap"]]
tabela['ROIC'] = tabela['ROIC'].str.replace("%", "")
tabela['ROIC'] = tabela['ROIC'].astype(str).replace('.', '')
tabela['ROIC'] = tabela['ROIC'].str.replace(".", "")
tabela['ROIC'] = tabela['ROIC'].str.replace(",", ".")
tabela['ROIC'] = tabela['ROIC'].astype(float)
```

```python
tabela['Volume'] = tabela['Volume'].str.replace('.', '')
tabela['Volume'] = tabela['Volume'].str.replace(".", "")
tabela['Volume'] = tabela['Volume'].replace(",", ".")
tabela['Volume'] = tabela['Volume'].str.replace(",", ".")
tabela['Volume'] = tabela['Volume'].replace(",", ".")
tabela['Volume'] = tabela['Volume'].astype(float)
```

```python
tabela['Volume'] = tabela['Volume'].str.replace('.', '')
tabela['Volume'] = tabela['Volume'].str.replace(".", "")
tabela['Volume'] = tabela['Volume'].replace(",", ".")
tabela['Volume'] = tabela['Volume'].str.replace(",", ".")
tabela['Volume'] = tabela['Volume'].replace(",", ".")
tabela['Volume'] = tabela['Volume'].astype(float)
```

```python
tabela['EV/EBIT'] = tabela['EV/EBIT'].astype(float)
tabela = tabela[tabela['Volume'] > 1000000]
tabela = tabela[tabela['EV/EBIT'] > 0]
tabela = tabela[tabela['ROIC'] > 0]

tabela['ranking_ev_ebit'] = tabela['EV/EBIT'].rank(ascending = True)
tabela['ranking_roic'] = tabela['ROIC'].rank(ascending = False)
tabela['ranking_total'] = tabela['ranking_ev_ebit'] + tabela['ranking_roic']
```
